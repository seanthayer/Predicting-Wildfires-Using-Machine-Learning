```python
df_bias = pd.read_csv("./stats/bias_acc.csv", sep = ',', header = 0)
df_bias = df_bias.set_index("Bias")
```

```python
plot_line_bias_mean = sns.lineplot(data = df_bias, x = "Bias", y = "Mean", color = "dodgerblue")
plot_line_bias_mean.set_title("One-shot Logistic Regression Mean Prediction Accuracy")
plot_line_bias_mean.set_xlabel("Bias")
plot_line_bias_mean.set_ylabel("Percentage");
```

```python
plot_line_bias_std = sns.lineplot(data = df_bias, x = "Bias", y = "Std", color = "dodgerblue")
plot_line_bias_std.set_title("One-shot Logistic Regression St.D. Prediction Accuracy")
plot_line_bias_std.set_xlabel("Bias")
plot_line_bias_std.set_ylabel("Percentage");
```
